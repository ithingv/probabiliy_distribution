# 감마 분포(Gamma distribution)

----
### 감마 함수 (Gamma function)

감마 분포에 알아보기 전에 먼저 감마 함수를 알아보자. 감마 함수는 아래 공식과 같이 정의되는 함수인데 이 감마 함수는 수학자들이 팩토리얼 $n! = n(n-1)(n-2)\cdots$을 자연수에만 한정해서 적용하지 않고, 더 큰 수 체계에도 적용할 수 있는 방법을 고민하다가 나온 함수라고 한다. 즉 복소수 범위까지 일반화 된 팩토리얼 인 것이다. 

<center>
$\large \Gamma \left( \alpha \right) = \int\limits_0^\infty {t^{\alpha - 1} e^{ - t} dt}$
</center>

이 감마함수는 아래 세가지 식과 같이 계산된다. 
>>
- $\Gamma(\alpha) = (\alpha-1)\Gamma(\alpha-1),$ $(\alpha > 1)$
- $\Gamma(\alpha) = (\alpha-1)!$
- $\Gamma(\frac{1}{2}) = \sqrt{\pi}$

## 감마 분포 (Gamma Distribution)

---

감마분포의 확률밀도함수는 다음과 같다.

<center>
 X ~ $Gamma(\alpha, \beta)$

$ \large f_X(x) = \frac{1}{\beta^{\alpha}\Gamma(\alpha)}x^{\alpha-1}e^{-\frac{x}{\beta}}$ ,  $ \small (x, \alpha, \beta, > 0)$

$(\alpha: \text {shape parameter}, \beta: \text{scale parameter})$
</center>


감마 분포는 $\alpha$개의 사건이 일어날 때까지 걸리는 대기 시간에 대한 분포라고 한다. 지수 분포는 사건이 1번 일어날 때 까지 걸리는 시간에 대한 것이었다. 바로 감마분포는 지수 분포를 한번의 사건이 아닌 여러 개의 사건으로 확장한 것이라고 할 수 있다.
