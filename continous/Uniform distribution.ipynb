# 균일 분포 (Uniform distribution)

균일분포는 정해진 범위에서 모든 확률이 균일한 분포를 의미한다. 가장 간단하면서도 응용도 많이 되는 분포이다. 범위가 0에서 1사이일 때는 하나의 지시 함수(indicator function)으로도 이용될 수 있다. 사실 균일 분포는 연속형 분포도 될 수 있지만 이산형 분포 버전도 있다. 이상현 균일 분포는 마찬가지로 가능한 $X$의 모든 값 {$x_1, x_2, \cdots, x_n$}에서 균일한 확률값을 가지는 분포일 것이다. 이 노트북에서는 연속형 균일 분포를 중점적으로 설명하려한다. 연속형 균일 분포의 확률밀도함수는 다음과 같이 나타낼 수 있다.

<center>
 X ~ $Uniform(a, b)$

$\large f_{X}(x) = \frac{1}{b-a}$ , $a < x < b$ 
</center>
