# 음이항 분포 (Negative Binomial distribution)

---
기하분포와 똑같은 조건에서 어떤 확률 변수 X를 r번째 성공을 얻을 때까지 걸리는 시행횟수라고 해보자. 이때의 X를 음이항확률변수라고 하고 분포를 음이항분포라고 한다. $(i.e. X $~$ Negative Binomial(r, p))$ 이때 r=1인 경우를 기하분포라 할 수 있고 음이항분포는 기하분포를 일반화한 것이라고 할 수 있다.

<center>
$X $~$NegativeBinomial(r,p)$

$f_{X}(x) = \binom{x-1}{r-1}p^{r}(1-p)^{x-r}$ , $x = r,r+1,\cdots$ 
</center>

음이항 분포는 '음'이항분포라는 이름을 가지고 있는데, 이름에서 음이항분포는 이항분포와 반대 개념이라는 것을 유추할 수 있다. 전체 시행 횟수를 n 성공횟수를 r이라 하고 X ~$ NB(r,p)$,  Y ~$B(n, p)$이라 할 때, 실제로 확률변수 X와 Y는 다음의 관계를 가지고 있다. 

<center>
$P(X≤n)=P(Y≥r)$
</center>

$P(X≤n)$ : $r$개의 성공을 거두는데 $n$ 이하의 시행 횟수가 필요한 사건$

$P(Y≥r)$ : $n$번의 시행에서 $r$개 이상의 성공을 거두는 사건

다시말해서, $r$개의 성공을 거두는데 $n$ 이하의 시행 횟수가 필요했다면, $n$번의 시행에서 r번 성공했거나 그 이상의 성공을 했을 수 있는 것이다. **즉, 음이항 분포에서는 시행 횟수 n이 확률변수이고 성공 횟수 r이 고정되어 있는 반면, 이항분포에서는 시행 횟수 n이 고정되어 있고 성공 횟수 r이 확률변수인 것입니다. 둘의 기준점이 다르다고 할 수 있겠네요!**






### 음이항분포의 포아송 분포로의 근사
---
한편, 음이항분포는 포아송 분포로 수렴된다는 특징이 있다. 이항 분포는 포아송 분포로 수렴된다는 것을 보였는데. 역시나 음이항 분포도 마찬가지로 포아송 분포로 수렴한다.  확률변수 X∼$NB(r,p)$이고, $r→∞$, $p→1$, $r(1−p)→λ$일 때, $ Poisson(λ)$에 수렴하게 됩니다
