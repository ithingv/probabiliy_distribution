# 초기하 분포 (Hypergeometric Distribution)

---
N명의 구성원으로 이루어진 어떤 모집단이 존재하고, 이 모집단이 두 그룹으로 나누어진다고 가정하다. 첫번째 그룹에는 M명의 구성원이 있다고 하면 다른 그룹에는 N-M명의 구성원이 존재할 것이다. 이 때, 초기하 확률변수 X는 전체 모집단에서 K개의 샘플을 비복원으로 뽑을 때, K개 샘플 중에서 첫번째 그룹에 해당하는 샘플 수를 의미한다.

<center>
 X ~ $Hypergeometric(N, M, K)$

$f_{X}(x) = \frac{\binom{M}{x}\binom{K-M}{K-x}}{\binom{N}{K}}$ , $(x = max(0, M -N + K),\cdots, min(K, M))$ 
</center>

다시 강조하면 초기하분포에서는 샘플을 **비복원추출**하는 것이다. 그런데 위와 똑같은 상황에서 샘플을 복원추출하면 **이항분포**에 해당한다. 똑같은 상황에서 샘플을 복원추출하는 개념은 이항분포, 비복원추출하는 개념은 초기하분포라고 정리할 수 있을 것이다.
